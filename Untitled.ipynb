## Data Cleaning

```python
#Making a copy of the original dataframe
df1 = df.copy(deep=True)
df1.head()
```

```python
#Checking the columns of the dataset
df1.columns
```

```python
#Striping off spaces and changing all letters to lowercase 
df1.columns= df1.columns.str.lower().str.strip().str.replace(".", "_")
df1.columns
```

```python
#Checking for duplicates
df1.duplicated().sum()
```

```python
#Checking for the value counts of the categorical columns
for col in df1.select_dtypes(include="O"):
    val_cnt= df1[col].value_counts(dropna=False)
    print(f"{val_cnt}", '\n' )
```

```python
#Drop the unnecessary columns
df1.drop(columns=["investigation_type", "accident_number", "airport_code", "airport_name", "registration_number", 
                  "publication_date", "report_status", "air_carrier"], axis=1, inplace=True)
df1.columns
```

```python
#To check if the columns dropped
df1.shape
```

```python
#Fill the null values in categorical columns with "Unknown"
def fill_nans_with_unknown(df1):
    """
    Fills all NaN values in categorical columns with the string "Unknown"
    """
    for col in df1.select_dtypes(include="O").columns:
        df1[col] = df1[col].fillna("Unknown")
    return df1

fill_nans_with_unknown(df1)
```

```python
#Checking whether there are still categoriacl columns with null values
df1.isna().sum()
```

```python
#Checking the value counts of the categorical columns to see of there are repeated values that need to be replaced
for col in df1.select_dtypes(include="O"):
    val_cnt= df1[col].value_counts(dropna=False)
    print(f"{val_cnt}", '\n' )
```

```python
#A function to help in standardization of the values in the columns
def clean_string_columns(df):
    """
    Strips whitespace, lowers then title-cases all string columns in a DataFrame.
    """
    for col in df.select_dtypes(include="O").columns:
        df[col] = df[col].astype(str).str.strip().str.lower().str.title()
    return df

clean_string_columns(df1)
```

```python
#Checking the value counts of the categorical columns after standardization
for col in df1.select_dtypes(include="O"):
    val_cnt= df1[col].value_counts(dropna=False)
    print(f"{val_cnt}", '\n' )
```

```python
#Replacing the name of a value because the value is similar to the other one
df1["purpose_of_flight"]= df1["purpose_of_flight"].str.replace("Air Race/Show", "Air Race Show")
df1["purpose_of_flight"].value_counts()
```

```python
#Replacing the value "unk" with "Unknown" all through the dataset
df1.replace("Unk", "Unknown", inplace=True)
df1["schedule"].value_counts()
```

```python
#Grouping the values in the injury_severity column and adding a new column, to simplify our analysis
def simplify_injury_severity(severity):
    if pd.isna(severity):
        return "Unknown"
    severity = severity.strip().lower()
    
    if severity.startswith("fatal"):
        return "Fatal"
    elif severity == "non-fatal":
        return "Non-Fatal"
    elif severity == "serious":
        return "Serious"
    elif severity == "minor":
        return "Minor"
    elif severity == "none":
        return "None"
    elif severity == "unavailable":
        return "Unknown"
    else:
        return "Other"
df1["injury_severity_grouped"] = df1["injury_severity"].apply(simplify_injury_severity)
df1["injury_severity_grouped"].value_counts()
```

```python
#Checking for duplicates in the new dataset after cleaning
df1.duplicated().sum()
```
