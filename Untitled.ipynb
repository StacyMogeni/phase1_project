After carefully examining the columns, I have decided to fill the null values in the categorical columns with "Unknown" in order to avoid bias, because using the mode might skew the data toward the most frequent category, which may not reflect reality, and to also prevent misleading analysis. For this, I am going to create a function that I will use to help me fill every null value in each column with "Unknown", hence making my work easier, instead of repeating the process over and over again.

```python
#Fill the null values in categorical columns with "Unknown"
def fill_nans_with_unknown(df1):
    """
    Fills all NaN values in categorical columns with the string "Unknown"
    """
    for col in df1.select_dtypes(include="O").columns:
        df1[col] = df1[col].fillna("Unknown")
    return df1

fill_nans_with_unknown(df1)
```

Checking the number of rows with the value "Unavailable" so that we can replace them with "Unknown", because "Unavailable" is a vague term, and "Unknown" is preferred here. This is because it clearly communicates that the value is missing or not recorded, and it helps avoid category clutter caused by multiple labels that mean the same thing; hence makes the data cleaner and simplifies analysis and visualization

```python
#Check if there are any rows with the value "Unavailable" in the injury_severity column
df1[df1["injury_severity"]== "Unavailable"]
```

To standardize all string values in the dataset, we will strip extra spaces, convert to lowercase, and apply title case as this helps ensure consistency in categories, making the data cleaner and avoiding issues like duplicate categories that differ only in formatting. We are going to use a function to make our work easier so that we don't have to repeat the same process over and over again.

```python
#A function to help in standardization of the values in the columns
def clean_string_columns(df):
    """
    Strips whitespace, lowers then title-cases all string columns in a DataFrame.
    """
    for col in df.select_dtypes(include="O").columns:
        df[col] = df[col].astype(str).str.strip().str.lower().str.title()
    return df

clean_string_columns(df1)
```

## Feature Engineering

I am going to use feature engineering to create new columns that may be of help in my exploratory data analysis.

The first column I am going to simplify is the Injury Severity column which has a lot of fatal values that can be grouped into one. This grouping will help to simplify the categories and make patterns easier to understand, therefore making the data more meaningful and analysis-ready. I will group the categories and name the new column Grouped Injury Severity.

```python
#Grouping the values in the injury_severity column and adding a new column
def simplify_injury_severity(severity):
    if pd.isna(severity):
        return "Unknown"
    severity = severity.strip().lower()
    
    if severity.startswith("fatal"):
        return "Fatal"
    elif severity == "non-fatal":
        return "Non-Fatal"
    elif severity == "serious":
        return "Serious"
    elif severity == "minor":
        return "Minor"
    elif severity == "none":
        return "None"
    elif severity == "unknown":
        return "Unknown"
    elif severity == "unavailable":
        return "Unavailable"
    else:
        return "Other"
df1["grouped_injury_severity"] = df1["injury_severity"].apply(simplify_injury_severity)
df1["grouped_injury_severity"].value_counts()
```

The next column I am going to add is the Total Injured column which basically calculates the total number of people that were injured in each row by adding the total fatal, total minor and total serious injuries. This column might help us in the exploratory data analysis.

```python
#Adding a new column that gives us the total number of people that were injured
df1["total_injured"]= df1["total_fatal_injuries"] + df1["total_minor_injuries"] + df1["total_serious_injuries"]
df1.head()
```

Another column that i will add is the Total Onboard column that calculates the total number of people that were in an aircraft by adding all those that were injured together with the ones that were uninjured.

```python
#Adding a new column that gives us the count of total people that were onboard
df1["total_onboard"] = (df1["total_fatal_injuries"] + df1["total_serious_injuries"] + df1["total_minor_injuries"] 
                          + df1["total_uninjured"])
df1.head()
```

```python
#Checking if there are any rows that have 0 as a value in the total aboard
df1[df1["total_onboard"] == 0]
```

#### An Observation:

From the above, we can see that there are rows that actually have 0 as the total number of people that were onboard, and that does not make much sense because there is no way an aircraft can have 0 people onboard. It atleast has to have a pilot, so the minimum number should at least be 1, which is not the case here. This issue likely arose during the imputation of missing values, where we used the median to fill in nulls. If the majority of entries were small aircraft with only one or two people, and if many missing values were in rows where the aircraft had few or no passengers, then the median could easily be 0. So, to fix this to make it more realistic, we are going to replace all the 0 values in the Total Onboard column with the median of that column.

```python
#Checking the median of the column
median_pass= df1["total_onboard"].median()
median_pass
```

```python
#Replacing the 0 values, with the median of that column
df1['total_onboard'] = df1['total_onboard'].replace(0, 2)
df1[df1["total_onboard"] == 0]
```

I also opted to add a new column that calculates the rate of injury in each row. It takes the value of the total number of people injured, and divides it by the total number of people onboard.This will be a very helpful column in exploratory data analysis.

```python
#Adding a new "injury_rate" column
df1['injury_rate'] = df1['total_injured'] / df1['total_onboard']
df1.head()
```

### Bivariate Analysis

```python
# Only keep the most common makes, cause they are too many
top_makes = df2['make'].value_counts().head(10).index

# Filter to those makes
top_df2 = df2[df2['make'].isin(top_makes)]

# Group by "make" and get average injury rate
avg_injury_rate = top_df2.groupby('make')['injury_rate'].mean().sort_values()

plt.figure(figsize=(9, 6))
sns.barplot(y=avg_injury_rate.values, x=avg_injury_rate.index, color="#BE5B50")
plt.ylabel('Average Injury Rate')
plt.xlabel('Aircraft Make')
plt.title('Average Injury Rate per Aircraft Make')
plt.show()
```

```python
# Calculate average injury rate per flight purpose
avg_injury_rate = df2.groupby('aircraft_category')['injury_rate'].mean().sort_values()

plt.figure(figsize=(12, 6))
sns.barplot(x=avg_injury_rate.index, y=avg_injury_rate.values, color="#A4B465")
plt.title('Average Injury Rate by Aircraft Category')
plt.ylabel('Average Injury Rate')
plt.xlabel('Aircraft Category')
plt.xticks(rotation=90)
plt.show()
```

```python
# Calculate average "injury_rate" by "amateur_built" 
avg_injury_rate = df2.groupby('amateur_built')['injury_rate'].mean().sort_values()

plt.figure(figsize=(6, 4))
sns.barplot(x=avg_injury_rate.index, y=avg_injury_rate.values, color="#FBDB93")
plt.title('Average Injury Rate: Amateur-Built vs Factory-Built')
plt.xlabel('Amateur Built')
plt.ylabel('Average Injury Rate')
plt.show()
```

### Multivariate Analysis

```python
#Get average injury rate per make and category
grouped = df2.groupby(['aircraft_category', 'make'])['injury_rate'].mean().reset_index()

#Filter to top 10 most common makes for readability
top_makes = df2['make'].value_counts().head(10).index
filtered = grouped[grouped['make'].isin(top_makes)]

plt.figure(figsize=(12, 6))
sns.barplot(data=filtered, x='make', y='injury_rate', hue='aircraft_category', palette='Set2')
plt.title('Average Injury Rate by Aircraft Make and Category')
plt.xlabel('Aircraft Make')
plt.ylabel('Average Injury Rate')
plt.xticks(rotation=45)
plt.legend(title='Aircraft Category')
plt.show()
```

```python
#Group and calculate average injury rate
grouped = df2.groupby(['purpose_of_flight', 'aircraft_category'])['injury_rate'].mean().reset_index()

#Filter to top 6 flight purposes for readability
top_purposes = df2['purpose_of_flight'].value_counts().head(6).index
filtered = grouped[grouped['purpose_of_flight'].isin(top_purposes)]

plt.figure(figsize=(12, 6))
sns.barplot(data=filtered, x='purpose_of_flight', y='injury_rate', hue='aircraft_category', palette='Set2')
plt.title('Average Injury Rate by Purpose of Flight and Aircraft Category')
plt.xlabel('Purpose of Flight')
plt.ylabel('Average Injury Rate')
plt.xticks(rotation=45)
plt.legend(title='Aircraft Category')
plt.tight_layout()
plt.show()
```

```python
#Calculate average injury rate grouped by Amateur Built and Make
grouped = df2.groupby(['amateur_built', 'make'])['injury_rate'].mean().reset_index()

#Focus on top 10 most frequent makes for clearer visualization
top_makes = df2['make'].value_counts().head(10).index
filtered = grouped[grouped['make'].isin(top_makes)]

plt.figure(figsize=(12, 6))
sns.barplot(data=filtered, x='make', y='injury_rate', hue='amateur_built',palette='pastel')
plt.title('Average Injury Rate by Aircraft Make and Amateur-Built Status')
plt.xlabel('Aircraft Make')
plt.ylabel('Average Injury Rate')
plt.xticks(rotation=45)
plt.legend(title='Amateur Built')
plt.tight_layout()
plt.show()
```
