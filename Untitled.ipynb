```python
#Fill the null values in categorical columns with "Unknown"
def fill_nans_with_unknown(df1):
    """
    Fills all NaN values in categorical columns with the string "Unknown"
    """
    for col in df1.select_dtypes(include="O").columns:
        df1[col] = df1[col].fillna("Unknown")
    return df1

fill_nans_with_unknown(df1)
```

```python
#A function to help in standardization of the values in the columns
def clean_string_columns(df):
    """
    Strips whitespace, lowers then title-cases all string columns in a DataFrame.
    """
    for col in df.select_dtypes(include="O").columns:
        df[col] = df[col].astype(str).str.strip().str.lower().str.title()
    return df

clean_string_columns(df1)
```

## Feature Engineering

```python
#Grouping the values in the injury_severity column and adding a new column, to simplify our analysis
def simplify_injury_severity(severity):
    if pd.isna(severity):
        return "Unknown"
    severity = severity.strip().lower()
    
    if severity.startswith("fatal"):
        return "Fatal"
    elif severity == "non-fatal":
        return "Non-Fatal"
    elif severity == "serious":
        return "Serious"
    elif severity == "minor":
        return "Minor"
    elif severity == "none":
        return "None"
    elif severity == "unknown":
        return "Unknown"
    elif severity == "unavailable":
        return "Unavailable"
    else:
        return "Other"
df1["grouped_injury_severity"] = df1["injury_severity"].apply(simplify_injury_severity)
df1["grouped_injury_severity"].value_counts()
```

```python
df1["total_injured"]= df1["total_fatal_injuries"] + df1["total_minor_injuries"] + df1["total_serious_injuries"]
df1.head()
```

```python
#Adding a new column that gives us the count of total people that were onboard
df1["total_onboard"] = (df1["total_fatal_injuries"] + df1["total_serious_injuries"] + df1["total_minor_injuries"] 
                          + df1["total_uninjured"])
df1.head(10)
```

```python
df1[df1["total_onboard"] == 0]
```

```python
#Adding a new "injury_rate" column
df1['injury_rate'] = df1['total_injured'] / df1['total_onboard']
df1.head()
```

### Bivariate Analysis

```python
# Creating a heatmap using the correlation of the two columns
corr = df1[['total_onboard', 'total_injured']].corr()

plt.figure(figsize=(6, 5))
sns.heatmap(corr, annot=True, cmap='summer', fmt='.2f')
plt.title('Correlation: Total Aboard vs Total Injured')
plt.show()
```

```python
#Ignore the rows with the value 0 in the column "total_onboard"
clean_df1 = df1[(df1['total_onboard'] > 0) & (df1['injury_rate'].notna())]

# Only keep the most common makes, cause they are too many
top_makes = clean_df1['make'].value_counts().head(10).index

# Filter to those makes
top_df1 = clean_df1[clean_df1['make'].isin(top_makes)]

# Group by "make" and get average injury rate
avg_injury_rate = top_df1.groupby('make')['injury_rate'].mean().sort_values()

plt.figure(figsize=(9, 6))
sns.barplot(y=avg_injury_rate.values, x=avg_injury_rate.index, color="#129990")
plt.ylabel('Average Injury Rate')
plt.xlabel('Aircraft Make')
plt.title('Average Injury Rate per Aircraft Make')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df1, x='aircraft_category', y='injury_rate', palette='Set2')
plt.title('Injury Rate by Aircraft Category')
plt.ylabel('Injury Rate')
plt.xticks(rotation=45)
plt.show()
```
